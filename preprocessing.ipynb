```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt 
import os
import cv2

```

```python
DIR = 'D:/projects/Scenic images/dataset/seg_train/seg_train'
DIR2 = 'D:/projects/Scenic images/dataset/seg_test/seg_test'
CATEGORIES = ['buildings', 'forest','glacier', 'mountain', 'sea', 'street']
```

```python
for category in CATEGORIES:
    path = os.path.join(DIR, category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap = 'gray')
        plt.show()
        break
```

```python
SIZE = img_array.shape[0]
print(SIZE)
# IMG_SIZE = 50
new_array = cv2.resize(img_array,(SIZE, SIZE))
```

```python
training_set = []
test_set = []
def create_sets():
    for category in CATEGORIES:
        path = os.path.join(DIR, category)
        path2 = os.path.join(DIR2, category)
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array,(SIZE, SIZE))
                training_set.append([new_array, class_num])
            except Exception as e:
                pass
        for img2 in os.listdir(path2):
            try:
                img_array2 = cv2.imread(os.path.join(path2, img2), cv2.IMREAD_GRAYSCALE)
                new_array2 = cv2.resize(img_array2,(SIZE, SIZE))
                test_set.append([new_array2, class_num])
            except Exception as e:
                pass

create_sets()
```

```python
import random
random.shuffle(training_set)
random.shuffle(test_set)
```

```python
print(len(training_set))
for i in training_set[:10]:
    print(i)
```

```python
Xtrain= []
Xtest = []

ytrain = []
ytest = []
```

```python
for features, labels in training_set:
    Xtrain.append(features)
    ytrain.append(labels)
for features, labels in test_set:
    Xtest.append(features)
    ytest.append(labels)
Xtrain = np.array(Xtrain).reshape(-1, SIZE, SIZE, 1)
ytrain = np.array(ytrain)
Xtest = np.array(Xtest).reshape(-1, SIZE, SIZE, 1)
ytest = np.array(ytest)
```

```python
import pickle
pickle_out = open("Xtrain.pickle","wb")
pickle.dump(Xtrain,pickle_out)
pickle_out.close()

pickle_out = open("ytrain.pickle","wb")
pickle.dump(ytrain,pickle_out)
pickle_out.close()

pickle_out = open("Xtest.pickle","wb")
pickle.dump(Xtest,pickle_out)
pickle_out.close()

pickle_out = open("ytest.pickle","wb")
pickle.dump(ytest,pickle_out)
pickle_out.close()
```
